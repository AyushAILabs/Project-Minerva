### Tokenize the text using NLTK(Natural Language Toolkit) in python 

```python
import nltk
nltk.download('punkt') # Install the “punkt” tokenizer models needed for sentence and word tokenization.
```

```python
from nltk import sent_tokenize
# sent_tokenize() function is used to split a text into sentences.
text = " Virat Kohli is a GOAT. He is the best batsman in the world."
sentences = sent_tokenize(text)
sentences # list of sentences in the text
```

```python
from nltk.tokenize import word_tokenize
# word_tokenize() function is used to split a text into words.
text = " Virat Kohli is a GOAT. He is the best batsman in the world."
words = word_tokenize(text)
words # list of words in the text
```

```python
from nltk.tokenize import WordPunctTokenizer
# WordPunctTokenizer() function is used to split a text into words and punctuation.
tokenizer = WordPunctTokenizer()
text = " Virat Kohli is a GOAT. He is the best batsman in the world. Don't ever mess with him!"
tokens = tokenizer.tokenize(text)
tokens # list of words and punctuation in the text
```

```python
from nltk.tokenize import TreebankWordTokenizer
# TreebankWordTokenizer() function is used to split a text into words and punctuation using the Penn Treebank tokenizer.
# this tokenizer is more intelligent and handles contractions and punctuation better than WordPunctTokenizer.
tokenizer = TreebankWordTokenizer()
text = " Virat Kohli is a GOAT. He is the best batsman in the world. Don't ever mess with him!"
tokens = tokenizer.tokenize(text)
tokens # list of words and punctuation in the text
```

```python
from nltk.tokenize import RegexpTokenizer
# RegexpTokenizer() function is used to split a text into words using a regular expression pattern.
# In this example, we are using the pattern '\w+' which matches sequences of word characters (letters, digits, and underscores).    
tokenizer = RegexpTokenizer(r'\w+')
text = " Virat Kohli is a GOAT. He is the best batsman in the world. Don't ever mess with him!"
tokens = tokenizer.tokenize(text)
tokens # list of words in the text (punctuation is removed)
```
